```python
!pip install ultralytics
import cv2
from ultralytics import YOLO


model = YOLO("yolo11n.pt")


video_source = "https://youtu.be/6_EZPZ2dbG0?si=TOvRXdfl7YYOfElL"
fps = 30


results = model(video_source, stream=True, verbose=False, device=0, imgsz=320)


for i, frame_results in enumerate(results):
    if i % (fps*2) == 0:
        seconds = i // fps
        target_ids = [2, 3, 5, 7]
        count = sum(1 for box in frame_results.boxes if int(box.cls[0]) in target_ids)
        print(f"Time: {seconds}s | total vehicles = {count}")
```
